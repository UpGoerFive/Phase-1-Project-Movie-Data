**Converting Column Values to Floats**

```python
tn_df['domestic_gross'] = tn_df['domestic_gross'].str.replace('$', '')
```

```python
tn_df['domestic_gross'] = tn_df['domestic_gross'].str.replace(',', '')
```

```python
tn_df['domestic_gross'] = tn_df['domestic_gross'].astype('int64')
```

```python
tn_df['worldwide_gross'] = tn_df['worldwide_gross'].str.replace('$', '')
```

```python
tn_df['worldwide_gross'] = tn_df['worldwide_gross'].str.replace(',', '')
```

```python
tn_df['worldwide_gross'] = tn_df['worldwide_gross'].astype('int64')
```

```python
tn_df['production_budget'] = tn_df['production_budget'].str.replace('$', '')
```

```python
tn_df['production_budget'] = tn_df['production_budget'].str.replace(',', '')
```

```python
tn_df['production_budget'] = tn_df['production_budget'].astype('int64')
```
